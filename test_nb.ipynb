```python
store_path = Path.cwd()
url = "file:///Users/u1166368/scratch/virtualizarr/ocean-3d-pot_rho_0-1-monthly-mean-ym_1900_01.nc" 

store = LocalStore(prefix=store_path)
registry = ObjectStoreRegistry({url: store})

vds = vz.open_virtual_dataset(
    url,
    registry=registry,
    parser=HDFParser(),
    decode_times=False
    )
```

```python
import xarray as xr
import icechunk

if Path("/Users/u1166368/scratch/virtualizarr/zarr_store").exists():
    import shutil
    shutil.rmtree("/Users/u1166368/scratch/virtualizarr/zarr_store")

# Create a new repository instance with virtual chunk container permissions for reading
config = icechunk.RepositoryConfig.default()
config.set_virtual_chunk_container(
    icechunk.VirtualChunkContainer(
        url_prefix="file:///Users/u1166368/scratch/virtualizarr/",
        store=icechunk.local_filesystem_store("/Users/u1166368/scratch/virtualizarr/zarr_store")
    ),
)

# Open the repository with the config that includes virtual chunk permissions
storage = icechunk.local_filesystem_storage("/Users/u1166368/scratch/virtualizarr/zarr_store")
repo = icechunk.Repository.create(storage, config)

# Write your virtual dataset to the repository
write_session = repo.writable_session("main")
vds.to_zarr(write_session.store)
write_session.commit("Initial commit of virtual dataset")

# Now you can read the data
xr.open_zarr(write_session.store, consolidated=False)
```
