```python
# Claudes bugfixes
import icechunk
from pathlib import Path

storage = icechunk.local_filesystem_storage(
    path=str(Path("~/cj877-atm-icechunk").expanduser())
)

config = icechunk.RepositoryConfig.default()
config.set_virtual_chunk_container(
    icechunk.VirtualChunkContainer(
        f"file://{store_root}/",
        icechunk.local_filesystem_store(store_root),
    )
)

# Must explicitly authorize the container (icechunk is safe-by-default and
# won't read virtual chunks unless the caller opts in). None = no credentials
# needed for local filesystem, but you still must pass it.
credentials = icechunk.containers_credentials({
    f"file://{store_root}/": None,
})

repo = icechunk.Repository.create(storage, config, credentials)

session = repo.writable_session("main")

combined_vds.vz.to_icechunk(session.store)

session.commit("Wrote to virtual icechunk store!")
```

Notes: 
- Kerchunk remote read occurs via Swift. Why is performance so much worse via ARE than remotely? This would be worth understanding.
- Icechunk does not have Swift support (AFAIK, will investigate further).
- Virtualisation is *always faster* than opening the raw files, although this would probably decrease when we move to smaller datasets (fewer files).
- Updating the catalog to work with icechunk would be a fair amount more work than updating to kerchunk, as we'll need to add the capabilities to intake-esm. However, this should probably be a long term goal for that library anyway.
- Updating to icechunk also lets us do some clever stuff with the catalog, in particular around checking that datastores are up to date (https://virtualizarr.readthedocs.io/en/stable/api/serialization.html#virtualizarr.accessor.VirtualiZarrDataTreeAccessor.to_icechunk). This would allow for more frequent rebuilds (ie. nightly).

# What we would like to do

- Support for rectilinear chunk grids is coming to [icechunk](https://github.com/earth-mover/icechunk/issues/1338#event-23657319028) and more generally [zarr](https://github.com/zarr-developers/zarr-python/pull/3534).
- This will unblock total virtualisation of the catalog, which will:
  1. Make opening any datasets in the catalog way quicker.
  2. Totally decouple the metadata and the data itself. The remote reads of icechunk & the kerchunk datastores are this: try to load the data itself (on gdata) and it fails. Implications:
        - Discoverability no longer requires being on Gadi, we can ship xarray data reprs to the interactive catalog (missing & very useful capability). See [Pangeo Catalog example](https://catalog.pangeo.io/browse/master/atmosphere/sam_ngaqua_qobs_eqx_3d/).
        - Way better performance - both due to less files being touched (lustre), cached concatenation (this *is* what virtualisation is).
- I personally think it's crucial that we try to keep as close to a single source of truth as possible with this. Currently we have two mirrors of the same catalog - `/g/data/xp65/` (`intake.cat.access_nri`) and in Nectar (interactive data catalogue). This is not really optimal since it's possible for them to get out of sync. If possible, the 'best' solution would be to move the interactive data catalog to Nirin, provided we can access it from the compute nodes (presently or request it from NCI). We then get rid of the `/g/data/xp65` version and/or use it as a build step, so we have a single catalog.
- The potential for this sort of issue/chicanery gets multiplied the more of this duplicate infra we have. Right now, we could
  1. Virtualise a dataset
  2. Serialise as kerchunk & as icechunk
  3. Read kerchunk remotely for interactive data catalogue and icechunk locally for `intake.cat.access_nri`.
- The number of datasets >> the number of catalogue files, so the potential for things to get out of sync is much higher. This will basically break our discovery/hand-holding pipeline, which is a really bad look for a toolchain designed to make life easier.
- So realistically, we need a single set of virtualised indices which both toolchains read.

# Blockers for Kerchunk

- Shocking performance on Gadi would need fixing.
- Rectilinear chunk grids are yet to be implemented in Zarr-Python, and I think Martin will need to wait for that before he can do the kerchunk implementation.

# Blockers for icechunk
- s3 credentials required. This could be worked around by having the interactive catalogue contain them, but that's far from secure.
- We could also have the interactive catalogue fetch the xarray repr's from a server running in Nirin which has access to credentials. This might be simpler.
- If we don't have anonymous s3 access, we'd also need to store credentials either in the xp65 conda environment or somewhere in the ACCESS-NRI Intake Catalog Package. Again, we do similar things already, but I think we should seriously consider making anonymous s3 access actually work, as it's simpler.
- We need to integrate it with intake-esm. This might not be too hard for a 'just fetch me the head' of `main` workflow. 
- 2.0 Release. Rectilinear chunk grids are implemented, we just need the feature to be released.

# Blockers for both
- Having a single source of truth for virtualised datasets requires being able to talk to the Nirin object storage from compute nodes, at the very least. There may be other complexities I'm not aware of.

___
# Cliff Notes
Most likely optimal strategy:
- Virtualise with icechunk, write virtual icechunk stores into Nirin Object storage.
- Have a server running in Nirin which serves the xarray reprs for these datasets by reading from the buckets sitting next to it for interactive catalogue.
- The catalog on Gadi just reads virtual datasets out of these buckets.
- Ideally, we move the catalog itself (parquet files) into one of these buckets too.
- Requires some care to ensure we don't mess everything up on a compute node. I think Andrew said there are precedents which might allow us to do this.
